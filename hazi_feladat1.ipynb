```sql
SELECT Count(*) AS "NULL adatok aránya a szall_cim oszlopban"
FROM rendeles
WHERE szall_cim is NULL

UNION

SELECT Count(*)
FROM rendeles
WHERE szall_cim is NOT NULL
```

```sql
SELECT szall_mod AS "Szállítási módok", COUNT(*) AS "Szállítások száma"
FROM rendeles
group by szall_mod
```

```sql
SELECT MAX(DATEDIFF(DAY,REND_DATUM,SZALL_DATUM)) AS "Legnagyobb kiszállítási idő",MIN(DATEDIFF(DAY,REND_DATUM,SZALL_DATUM)) AS "Legkisebb kiszállítási idő"
FROM rendeles
```
