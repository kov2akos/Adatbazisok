#### Labdarúgó bajnokság mérkőzéseinek nyilvántartása

Ez az adatbázis a labdarúgó mérkőzések, csapatok, játékosok, gólok és büntetések nyilvántartására készült.

#### Logikai modell

- Csapat(CsapatID, Név, Város)
- Játékos(JátékosID, Név, SzületésiDátum, Poszt, CsapatID)
- Mérkőzés(MérkőzésID, Dátum, Helyszín, HazaiCsapatID, VendégCsapatID, HazaiGól, VendégGól)
- Gól(GólID, MérkőzésID, JátékosID, Perc)
- Büntetés(BüntetésID, MérkőzésID, JátékosID, Típus, Perc)

#### Kapcsolatok
- 1:N kapcsolat a Csapat és a Játékos között
- 1:1 kapcsolat a Mérkőzés és a Csapat között, kétszer (HazaiCsapatID és VendegCsapatID idegen kulcsok)
- 1:N kapcsolat a Játékos és a Gól között
- 1:N kapcsolat a Mérkőzés és a Gól között
- 1:N kapcsolat a Játékos és a Büntetés között
- 1:N kapcsolat a Mérkőzés és a Büntetés között
